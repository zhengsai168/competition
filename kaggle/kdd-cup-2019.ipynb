```python
import os
import pandas as pd
import math
```

```python
data_path = r'E:\data\data_set_phase1'
test_queries_data_path = os.path.join(data_path,'test_queries.csv')
test_plans_data_path = os.path.join(data_path,'test_plans.csv')
df_test_queries = pd.read_csv(test_queries_data_path)
df_test_plans = pd.read_csv(test_plans_data_path)
df_test_queries.head()
```

```python
def change(s):
    if math.isnan(s):
        return -1
    return int(s)
df_test_queries['pid'] = df_test_queries.pid.apply(change)
df_test_queries['sid'] = df_test_queries.sid.apply(change)
df_test_plans['sid'] = df_test_plans.sid.apply(change)
df_test_queries[:20]
```
